```python
import matplotlib.pyplot as plt
%matplotlib inline
# if u want to change the given style use
plt.style.use('seaborn-deep') # every style has its own plotting style 
# print(plt.style.available)  # this to see all the style available
```

```python
X = [1,2,3,4,5,6,7,89,9.0,99,90] # we can use the same parameter in another plot in a given axis
Y = [22,11,33,44,55,66,66,77,88,88,66]

Y_1 = [2,1,3,4,5,6,63,37,38,88,66]
plt.plot(X,Y,color = "red",linestyle= '--',marker='.',label='first1') # (these string goes as the colour) & ('--' is for ----- types of line) 
plt.plot(X,Y_1,color= 'black',linestyle="-",marker='o',label='second') # ploted another plot this is default "-"
# to use label arg in the plot itself then u need to have the plt.legend() there
plt.title("Normal Plot Just For Fun")
plt.xlabel("X")
plt.ylabel("Y")
# we can pass the marker color and linestyle arg as a single string also
# we can change the linewidth with the ar linewidth itself 
# in order to get a particuar line over an another line just cut an paste the code of the lines code afterwards 



# inorder to show the label of thge line we can use plt.legend
plt.legend()            # it takes argument as a list [ ]
# but this not a good practice asthe plot will be labeled acording to there order in code

# to avoid this mistake just pass the label arg inthe plot function directly


# inorder to get a grid view we use
plt.grid(True) # it takes argument true or not but by default it is true
# if the pattern is not showing up correctly use  plt.tight_layout()  here just above the show


# we can save the plot in our current directory y using this
# plt.savefig("firstplot.png")  name of the plot and also the the directory in which we want to save it 


plt.show()# the show should always be at the end
```
